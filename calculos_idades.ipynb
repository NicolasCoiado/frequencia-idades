```python
import pandas as pd
import matplotlib.pyplot as plt

idades = [24, 23, 45, 33, 24, 33, 42, 38, 33, 45, 24, 33]

def calculaMedias(lista):
    df_lista = pd.DataFrame(lista)
    df_lista.columns = ["idades"]
    media_idades = int(df_lista["idades"].mean())  # Calcula a mÃ©dia das idades
    print("MÃ©dia das idades:", media_idades)  # Imprime a mÃ©dia das idades

def freqDados(lista):
    serie = pd.Series(lista)  # Converte a lista em uma Series
    valores_unicos = serie.unique()  # Usa .unique() na Series
    frequencia_valores = serie.value_counts()
    tracaGrafico(frequencia_valores)  # Chama a funÃ§Ã£o para traÃ§ar o grÃ¡fico

def tracaGrafico(frequencia_valores):
    # Gera o grÃ¡fico de barras
    plt.figure(figsize=(8, 6))
    frequencia_valores.plot(kind='bar', color='skyblue')
    plt.title("FrequÃªncia de Idades")
    plt.xlabel("Idades")
    plt.ylabel("FrequÃªncia")

    # Define os ticks do eixo y como 0, 1, 2, 3 e 4
    plt.yticks(ticks=[0, 1, 2, 3, 4])

    plt.show()


calculaMedias(idades)
freqDados(idades)
```
